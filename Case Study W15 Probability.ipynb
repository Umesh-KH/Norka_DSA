# Case Study on Probability for Data Science

## Problem Statement:

To make a suitable machine learning algorithm to predict if the mushroom is edible or poisonous (e or p) using the given dataset. 
(Along with other ML algorithms, Naïve Bayes’ Classifier should be applied). 
Also, if some data pre-processing is necessary do that as well.

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score,accuracy_score,classification_report,precision_score,recall_score
from sklearn.metrics import classification_report

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB,BernoulliNB
```

## Importing the dataset

```python
data = pd.read_csv('mushrooms.csv')
data
```

```python
print("The dimension of the given dataset is", data.shape)
```

```python
# Checking for null values
data.isna().sum()
```

#### There are no null values present in the dataset.

```python
# The concise summary of the dataset can be given as
data.info()
```

```python
# The descriptive statistics of the dataset is as follows:
data.describe()
```

## Label Encoding

```python
# The column "CLASS" is label encoded. The column has two unique elements - 'p' and 'e'.

leng = LabelEncoder()
data['class'] = leng.fit_transform(data['class'])
data
```

### Splitting the dataset

```python
# Splitting the dataset into features and target set.

x = data.drop(['class'],axis=1) #features
y = data['class'] #target

# One-hot encoding on features.

x = pd.get_dummies(x)

# Splitting the data set into training and testing data

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 42, test_size = 0.35)
```

## Modelling Building - ML Algorithms & Predictions

### 1. KNN Classification 

```python
acc = []
for i in (np.arange(3,12)):
    classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski')
    classifier.fit(x_train,y_train)
    y_pred = classifier.predict(x_test)
    acc_score = accuracy_score(y_test,y_pred)
    acc.append(acc_score)

print("Accuracy is: ", accuracy_score(y_test,y_pred))
print("Precision is: ", precision_score(y_test,y_pred, average = 'micro'))
print("Recall is: ", recall_score(y_test,y_pred,  average = 'micro'))
print("F1 score is: ", f1_score(y_test,y_pred,  average = 'micro'))
```

### 2. Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
detr = DecisionTreeClassifier()
detr.fit(x_train, y_train)
y_pred = detr.predict(x_test)

print("Accuracy is: ", accuracy_score(y_test,y_pred))
print("Precision is: ", precision_score(y_test,y_pred, average = 'micro'))
print("Recall is: ", recall_score(y_test,y_pred,  average = 'micro'))
print("F1 score is: ", f1_score(y_test,y_pred,  average = 'micro'))
```

### 3. Logistic Regression

```python
classifier = LogisticRegression(solver='liblinear',C=10.0,random_state = 0)
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)

print("Accuracy is: ", accuracy_score(y_test,y_pred))
print("Precision is: ", precision_score(y_test,y_pred, average = 'micro'))
print("Recall is: ", recall_score(y_test,y_pred,  average = 'micro'))
print("F1 score is: ", f1_score(y_test,y_pred,  average = 'micro'))
```

### 4. Random Forest Classification

```python
rafo=RandomForestClassifier()
rafo.fit(x_train,y_train)
y_pred=rafo.predict(x_test)

print("Accuracy is: ", accuracy_score(y_test,y_pred))
print("Precision is: ", precision_score(y_test,y_pred,  average = 'weighted'))
print("Recall is: ", recall_score(y_test,y_pred,  average = 'weighted'))
print("F1 score is: ", f1_score(y_test,y_pred, average = 'weighted'))
```

## 5. Naive Bayes Classification 

### Naive Bayes Gaussian Classification

```python
from sklearn.naive_bayes import GaussianNB
nbg = GaussianNB()
nbg.fit(x_train, y_train)
print("Test Accuracy: {}%\n".format(round(nbg.score(x_test, y_test)*100, 2)))

y_pred_nb = nbg.predict(x_test)
print("Accuracy is: ", accuracy_score(y_test,y_pred_nb))
print("Precision is: ", precision_score(y_test,y_pred_nb, average = 'micro'))
print("Recall is: ", recall_score(y_test,y_pred_nb,  average = 'micro'))
print("F1 score is: ", f1_score(y_test,y_pred_nb,  average = 'micro'))
```

### Naive Bayes Bernoulli Classification

```python
nbb = BernoulliNB()
nbb.fit(x_train,y_train)
y_pred = nbb.predict(x_test)

print("Accuracy is: ", accuracy_score(y_test,y_pred))
print("Precision is: ", precision_score(y_test,y_pred, average = 'micro'))
print("Recall is: ", recall_score(y_test,y_pred,  average = 'micro'))
print("F1 score is: ", f1_score(y_test,y_pred,  average = 'micro'))   
```

### It can be observed that all the ML algorithms used, have prediction accuracy of nearly 100%.
